```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yahooFinance
```

```python
sns.set_style('whitegrid') # background
get_ipython().magic(u'matplotlib inline')
```

```python
# Reading stock data from the internet using pandas
import pandas_datareader.data as web
```

```python
# Time stamps
from datetime import datetime
```

```python
# Float numbers
from __future__ import division
```

## Fetch Data

```python
# Stonks
stonks = ['AAPL', 'MSFT', 'DELL', 'CRWD', 'TM', 'NKE']
```

```python
# start and end date for data grab
end = datetime.now()
start = datetime(end.year - 1, end.month, end.day)
```

```python
stonk_data = {}

for stonk in stonks:
    stonk_data[stonk] = (yahooFinance.Ticker(stonk).history(start=start, end=end))
```

## Single Stonk Analysis

```python
stonk = stonks[0]
print(stonk)
sd = stonk_data[stonk]
print(sd)
```

### Daily closing price

```python
sd['Close'].plot(legend=True, figsize=(10,4))
plt.show
```

### Total daily volume

```python
sd['Volume'].plot(legend=True, figsize=(10,4))
plt.show()
```

```python
# Calculate the moving average of the stock
ma_day = [10,20,50]

for ma in ma_day:
    column_name = "MA for %s days"%(str(ma))

    sd[column_name] = stonk_data[stonks[0]]['Close'].rolling(window=ma).mean()
```

### Moving average of closing price

```python
sd[['Close','MA for 10 days','MA for 20 days','MA for 50 days']].plot(subplots=False,figsize=(10,4))
plt.show()
```

### Daily returns 

```python
# Analyse daily returns of stonk
sd['Daily Return'] = sd['Close'].pct_change()
sd['Daily Return'].plot(figsize=(10, 4), legend=True, linestyle='--', marker='o')
plt.show()
```

### Avg daily return of the last year

```python
sns.displot(sd['Daily Return'].dropna(), bins=100, color='red')
plt.show()
```
